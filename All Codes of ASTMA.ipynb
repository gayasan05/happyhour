```python
#Lab 1
import re
import nltk
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords

nltk.download("punkt_tab", quiet=True) 
nltk.download("stopwords", quiet=True) 

text = "Hello World! Year 2026" 

print("Original Text:") 
print(text) 
print("-" * 50) 

tokens = word_tokenize(text)
print("Tokens:") 
print(tokens)  
print("-" * 50)

stop_words = set(stopwords.words('english')) 
filtered_tokens = []

for word in tokens:

    word = word.lower() 

    if word.isalpha() and word not in stop_words:
        filtered_tokens.append(word)

print("Filtered tokens:")
print(filtered_tokens)

print("-" * 50) 
    
def is_english_script(text):
    
    pattern = re.compile(r'^[A-Za-z0-9\s.,!?\'-]+$')
    return bool(pattern.match(text))

if is_english_script(text): 
    print("Script Validation: Text is English (Latin script)") 
else:
    print("Script Validation: Text contains invalid script")
```

```python
#Lab 2
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

# Download required resources
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")

text = "Android is better than Apple"

print("=" * 60)
print("CORRECT NLP PREPROCESSING PIPELINE")
print("=" * 60)

# Step 1: Original Input
print("\n1. Original Input Text:")
print(text)

# Step 2: Tokenization
tokens = word_tokenize(text)

print("\n2. Tokenized Words:")
print(tokens)

# Step 3: POS Tagging (Before Stopword Removal)
tagged_tokens = pos_tag(tokens)

print("\n3. POS Tagging:")
print(tagged_tokens)

# Step 4: Stopword Removal
stop_words = set(stopwords.words("english"))

filtered_tokens = [
    (word, tag) for word, tag in tagged_tokens
    if word.lower() not in stop_words and word.isalpha()
]

print("\n4. After Stopword Removal:")
print(filtered_tokens)

# Step 5: Stemming
stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word) for word, tag in filtered_tokens
]

print("\n5. Stemmed Words:")
print(stemmed_words)

# Step 6: Convert POS tags for WordNet
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith("J"):
        return wordnet.ADJ
    elif treebank_tag.startswith("V"):
        return wordnet.VERB
    elif treebank_tag.startswith("N"):
        return wordnet.NOUN
    elif treebank_tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Step 7: Lemmatization
lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in filtered_tokens
]

print("\n6. Lemmatized Words:")
print(lemmatized_words)

print("\n" + "=" * 60)
print("PIPELINE COMPLETED")
print("=" * 60)
```

```python
#Lab 3
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
import pandas as pd

# Documents
d0 = "This is a sample notebook"
d1 = "This notebook is sample"
d2 = "This is another sample notebook"

docs = [d0, d1, d2]

# ---------------- TF-IDF ----------------
print("=" * 60)
print("TF-IDF ANALYSIS")
print("=" * 60)

tfidf = TfidfVectorizer()
result = tfidf.fit_transform(docs)

print("\nIDF Values:")
for word, val in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(f"{word:10} : {val:.4f}")

print("\nWord Indexes:")
print(tfidf.vocabulary_)

print("\nTF-IDF Matrix:")
print(result.toarray())

# ---------------- COUNT VECTOR ----------------
print("\n" + "=" * 60)
print("DOCUMENT TERM MATRIX")
print("=" * 60)

cv = CountVectorizer()
matrix = cv.fit_transform(docs)

dtm_df = pd.DataFrame(
    matrix.toarray(),
    columns=cv.get_feature_names_out(),
    index=["d0", "d1", "d2"]
)

print(dtm_df)

# ---------------- COSINE ----------------
print("\n" + "=" * 60)
print("COSINE SIMILARITY")
print("=" * 60)

cosine_sim = cosine_similarity(matrix)

cosine_df = pd.DataFrame(
    cosine_sim,
    index=["d0", "d1", "d2"],
    columns=["d0", "d1", "d2"]
)

print(cosine_df)

# ---------------- JACCARD ----------------
print("\n" + "=" * 60)
print("JACCARD SIMILARITY")
print("=" * 60)

binary_cv = CountVectorizer(binary=True, token_pattern=r"(?u)\b\w+\b")
X = binary_cv.fit_transform(docs)

jaccard_dist = pairwise_distances(X.toarray(), metric="jaccard")
jaccard_sim = 1 - jaccard_dist

jaccard_df = pd.DataFrame(
    jaccard_sim,
    index=["d0", "d1", "d2"],
    columns=["d0", "d1", "d2"]
)

print(jaccard_df)
```

```python
# lab4_sentimentanalysis_fianance
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report, accuracy_score

def analyze_finance_data():
    analyzer = SentimentIntensityAnalyzer()

    print("=" * 70)
    print("Loading Finance Dataset...")
    print("=" * 70)

    try:
        df = pd.read_csv("C:/Users/Mohit/Downloads/ASTMA_lab-main/ASTMA_lab-main/data/lab5_sentiment_finance.csv", encoding="latin-1").dropna()
    except FileNotFoundError:
        print("Error: lab5_sentiment_finance.csv not found.")
        return

    # Prediction Function
    def get_vader_label(text):
        score = analyzer.polarity_scores(str(text).lower())["compound"]

        if score >= 0.05:
            return "Positive"
        elif score <= -0.05:
            return "Negative"
        else:
            return "Neutral"

    print("Applying VADER Sentiment Analysis...")
    df["vader_prediction"] = df["text"].apply(get_vader_label)

    # Format true labels
    df["true_label"] = df["category"].str.capitalize()

    valid_df = df.dropna(subset=["true_label", "vader_prediction"])

    # Sample Predictions
    print("\n" + "=" * 70)
    print("Sample Predictions")
    print("=" * 70)

    for i, row in valid_df.head(5).iterrows():
        print(f"\nNews {i+1}:")
        print(f"Text       : {str(row['text'])[:120]}...")
        print(f"Predicted  : {row['vader_prediction']}")

    # Accuracy
    acc = accuracy_score(valid_df["true_label"], valid_df["vader_prediction"])

    print("\n" + "=" * 70)
    print("Model Performance")
    print("=" * 70)
    print(f"Accuracy : {acc:.4f}")

    print("\nClassification Report:")
    print(classification_report(valid_df["true_label"], valid_df["vader_prediction"]))

if __name__ == "__main__":
    analyze_finance_data()
```

```python
# lab7_pagerank_git
import pandas as pd
import networkx as nx

def run_lab7_github():
    print("Loading GitHub Dataset...")
    df = pd.read_csv("C:/Users/Mohit/Downloads/ASTMA_lab-main/ASTMA_lab-main/data/lab8_github_edges.csv")

    print("Building Directed Graph...")
    G = nx.from_pandas_edgelist(
        df,
        'source',
        'target',
        create_using=nx.DiGraph()
    )

    print(f"Nodes: {G.number_of_nodes()} | Edges: {G.number_of_edges()}")

    print("Computing PageRank (alpha=0.85)...")
    pagerank = nx.pagerank(G, alpha=0.85)

    sorted_pr = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)

    print("\n--- Top 10 GitHub Nodes by PageRank ---")
    for i, (page, rank) in enumerate(sorted_pr[:10]):
        print(f"Rank {i+1}: Node {page} = {rank:.6f}")

if __name__ == "__main__":
    run_lab7_github()
```

```python
# Lab8_Centrality_Fb
import pandas as pd
import networkx as nx

def run_lab8_facebook():
    print("Loading Facebook Dataset...")
    df = pd.read_csv("C:/Users/Mohit/Downloads/ASTMA_lab-main/ASTMA_lab-main/data/lab8_facebook_edges.csv")

    # Create Undirected Graph (best for Facebook friendships)
    print("Building Undirected Graph...")
    G = nx.from_pandas_edgelist(
        df,
        source="source",
        target="target",
        create_using=nx.Graph()
    )

    # Basic Info
    print("\nGraph Summary:")
    print("Total Nodes:", G.number_of_nodes())
    print("Total Edges:", G.number_of_edges())

    # Compute Degree Centrality
    print("\nComputing Degree Centrality...")
    degree = nx.degree_centrality(G)

    # Compute Betweenness Centrality
    print("Computing Betweenness Centrality (Approximation k=100)...")
    betweenness = nx.betweenness_centrality(G, k=100, seed=42)

    # Sort Top 5
    top_degree = sorted(degree.items(), key=lambda x: x[1], reverse=True)[:5]
    top_between = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:5]

    # Display Results
    print("\n--- Top 5 Most Connected Users (Degree Centrality) ---")
    for node, val in top_degree:
        print(f"Node {node}: {val:.5f}")

    print("\n--- Top 5 Best Bridge Users (Betweenness Centrality) ---")
    for node, val in top_between:
        print(f"Node {node}: {val:.5f}")

if __name__ == "__main__":
    run_lab8_facebook()
```
